### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
STR = ("MUFASA")
SET <- V = ("A", "E", "I", "O", "U")
SET <- FINAL = 0
FOR EACH CH IN STR DO
  IF CH IN V:
    SET T_FINAL = CH + FINAL
END FOR
RETURN T_FINAL
```

```python
str = ("mufasa")
str = str.lower()
vowels = ('a','e','e','o','u')
count = 0
for ch in str:
  if ch in vowels:
    count += 1
print(count)
```
